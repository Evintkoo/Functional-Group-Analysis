```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Datas/clear_encoded_data.csv")
df
```

```python
atom_weights ={
    "Br": 79.907,
    "C" : 12.01,
    "Cl": 35.45,
    "F": 18.99,
    "H": 1.0079,
    "I": 126.904,
    "N": 14.0068,
    "O": 15.99,
    "P": 30.97,
    "S": 32.062
}
```

```python
for cols in df.columns:
    df[cols] *= atom_weights[cols]
```

```python
weight_df = pd.DataFrame(df.T.sum(), columns=["weight"])
weight_df.to_csv("Datas/molecular_weights.csv", index=False)
weight_df
```
