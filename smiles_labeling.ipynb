```python
df = pd.read_csv("C:/Users/Evint/Documents/Projects/Functional-Group-Analysis/250k_rndm_zinc_drugs_clean_3.csv")
df['smiles'] = df['smiles'].apply(lambda s: s.replace('\n', ''))
df = df[df["SAS"] > 3.55].reset_index()
df.drop(["SAS", "logP", "qed", "index"], axis=1, inplace=True)
df["label"] = label_data["label"]
df
```

```python
total_sample = 150
meta_data = {}
for i in np.unique(list(df["label"])):
    cur_df = df[df["label"] == i]
    smiles_list = list(cur_df["smiles"])
    id_count = 0
    smiles_list = random.choices(smiles_list, k=total_sample)
    cluster_metadata = {}
    for smiles in smiles_list:
        m = Chem.MolFromSmiles(smiles)
        Draw.MolToFile(m, "Stats/molecules/cluster_{}/{}.png".format(i,id_count),size=(1080,1080))
        cluster_metadata[str(id_count)] = smiles
        id_count += 1
    meta_data["cluster_{}".format(i)] = cluster_metadata
meta_data
```

```python
json.dump(meta_data, open("Stats/molecules/smiles_metadata.json", "w"))
```
