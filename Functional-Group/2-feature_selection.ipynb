```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Datas/clear_encoded_data.csv")
df
```

```python
def variance(X : np.array) :
    mean = np.mean(X)
    sum = np.sum([(xi - mean)**2 for xi in X])
    return sum/len(X)
```

```python
var_list = [(cols, variance(df[cols])) for cols in df.columns]
filetered_columns = [i for i, j in var_list if j > 0.5]
filetered_columns
```

```python
df = df[filetered_columns]
df
```

```python
df.to_csv("Datas/selected_data.csv", index=False)
```
