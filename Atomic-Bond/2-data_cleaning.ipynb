```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("Datas/encoded_data.csv")
df.shape
```

```python
col = df.columns
col = [eval(i) for i in col]
col
df.columns = col
df.head()
```

```python
list_selected_col = list()
for a, b, c in col:
    if a == b:
        list_selected_col.append((a,b,c))
    elif ((a,b,c) not in list_selected_col) and ((b,a,c) not in list_selected_col):
        list_selected_col.append((a,b,c))
len(list_selected_col)
```

```python
x = 0
list_selected_col = list()
for a, b, c in df.columns:
    if ((a,b,c) in df.columns and (b, a, c) in df.columns) and (a != b):
        df[(a,b,c)] += df[(b,a,c)]
        x += 1
    if a == b:
        list_selected_col.append((a,b,c))
    elif (a, b, c) not in list_selected_col and (b, a, c) not in list_selected_col:
        list_selected_col.append((a,b,c))
new = df[list_selected_col]
new.head()
```

```python
new.to_csv("Datas/cleared_data.csv", index=False)
```
