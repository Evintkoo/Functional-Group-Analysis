```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statistics

def save_data(X : pd.DataFrame, path:str):
    X.to_csv(path+".csv")
    X.to_excel(path+".xlsx")
```

```python
df = pd.read_csv("Datas/cleared_data.csv")
df
```

```python
variance_data = list()
for columns in df.columns:
    if statistics.variance(df[columns]) > 1 :
        variance_data.append((columns, statistics.variance(df[columns])))
variance_data = np.transpose(variance_data)
variance_data
```

```python
var_df = pd.DataFrame(variance_data, columns=variance_data[0]).T
var_df.drop(0, axis=1, inplace=True)
var_df
```

```python
save_data(var_df, "Datas/Stats/variance_data")
```
