```python
import pysmiles
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import networkx as nx
from itertools import chain
from collections import Counter
import random
import logging

logging.disable(logging.CRITICAL)
```

```python
df = pd.read_csv("250k_rndm_zinc_drugs_clean_3.csv")
df['smiles'] = df['smiles'].apply(lambda s: s.replace('\n', ''))
df.head()
```

```python
df = df[df["SAS"] > 3.55]
smiles = df["smiles"].values
len(smiles)
```

```python
def unique_edges_list(smile):
    smile = smile
    mol = pysmiles.read_smiles(smile, explicit_hydrogen=True, reinterpret_aromatic=True)
    elements = dict(mol.nodes(data='element'))
    li = list(mol.edges(data='order'))
    edge_list = list()
    for a,b,c in li:
        edge_list.append((elements[a],elements[b],c))
    return set(edge_list)
```

```python
smile = smiles
mol = pysmiles.read_smiles(smile, explicit_hydrogen=True, reinterpret_aromatic=True)
mol
elements = dict(mol.nodes(data='element'))
li = list(mol.edges(data='order'))
edge_list = list()
for a,b,c in li:
    edge_list.append((elements[a],elements[b],c))
#count number of bond
count_edge = dict(Counter(chain(edge_list)))
```

```python
smiles = smiles
len(smiles)
```

```python
unique_edge_list = [unique_edges_list(sm) for sm in smiles]
a = [list(i) for i in unique_edge_list]
x = list()
for i in a:
    for j in i:
        x.append(j)
col = set(x)
len(col)
```

```python
list_df = list()
for smile in smiles:
    enc = list()
    mol = pysmiles.read_smiles(smile, explicit_hydrogen=True, reinterpret_aromatic=True)
    elements = dict(mol.nodes(data='element'))
    li = list(mol.edges(data='order'))
    edge_list = list()
    for a,b,c in li:
        edge_list.append((elements[a],elements[b],c))
    count_edge = dict(Counter(chain(edge_list)))
    #print(count_edge)
    for c in col:
        try: 
            enc.append(count_edge[c])
        except:
            enc.append(0)
    list_df.append(enc)
list_df
```

```python
new_df = pd.DataFrame(list_df, columns=list(col))
new_df
```
