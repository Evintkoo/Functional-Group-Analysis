```python
import os
import pandas as pd
import numpy as np

path = "results\clustering"
mean_df = []
std_df = []
dd_path = [os.path.join(path, i) + "\\decoded_data.csv" for i in os.listdir(path)]
for p in dd_path:
    df = pd.read_csv(p)
    new_df = pd.DataFrame(scaler.inverse_transform(df.values), columns = df.columns)
    
    mean_df.append(new_df.mean())
    std_df.append(new_df.std())
mean = pd.concat(mean_df, axis=1).T
std = pd.concat(std_df, axis=1).T 
```

```python
mean = mean.round(2)
mean = mean[mean > 0.05].dropna(axis=1)
mean
```
